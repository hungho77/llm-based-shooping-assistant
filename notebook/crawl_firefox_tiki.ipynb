```python
import numpy as np

url = f"https://tiki.vn/search?q=headphone+bluetooth&price={price_from}%2C{price_to}"
driver.get(url=url)
item_info = []

# page 1
all = driver.find_element(By.XPATH, "//div[@class='CatalogProducts__Wrapper-sc-1r8ct7c-0 fKlIZg']")
all_items = all.find_elements(By.XPATH, "div[@class='']")
print(len(all_items))
for i, item in enumerate(all_items):
    item_info.append(item.text)

n_page = 2
while n_page <= 50:
    next_page_url = driver.find_element(By.XPATH, "//a[@class='arrow undefined']").get_attribute('href')
    if next_page_url is None:
        break
    
    print(next_page_url)
    driver.get(next_page_url)
    
    sleep(np.random.randint(1,5))
    
    #get all items each page
    all = driver.find_element(By.XPATH, "//div[@class='CatalogProducts__Wrapper-sc-1r8ct7c-0 fKlIZg']")
    all_items = all.find_elements(By.XPATH, "div[@class='']")
    # print(len(all_items))
    for i, item in enumerate(all_items):
        item_info.append(item.text)
        
    sleep(np.random.randint(1,5))
    n_page += 1
    
    
    
```

```python
arr = [", ".join(item.replace("TÀI TRỢ\n", "").split("\n")) for item in item_info]
arr
```

```python

# Tên của tệp tin để lưu
file_name = 'all_items_tiki.txt'

# Mở tệp tin để ghi, 'w' để ghi mới hoặc ghi đè tệp tin đã có
with open(file_name, 'w') as file:
    # Duyệt qua từng phần tử trong mảng
    for item in arr:
        # Ghi từng phần tử vào file, kết thúc bằng dấu xuống dòng
        file.write(item + '\n\n')
```

```python
info_all_items = [item.replace("TÀI TRỢ\n", "") for item in item_info]
info_all_items = [item.split('\n') for item in info_all_items]
info_all_items
```

```python

series_price = []
series_sale = []
series_brand = []
series_name_item = []

for item in info_all_items:
    series_price.append(item[0])
    series_sale.append(item[1])
    series_brand.append(item[2])
    series_name_item.append(item[3])
   
    
```
